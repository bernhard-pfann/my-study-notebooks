```python
import numpy as np
import pandas as pd
from scipy.stats import expon, geom, poisson
import matplotlib.pyplot as plt
```

```python
lam=4
rng = np.arange(0, 20, 1)
exp_pdf = pd.Series(data=expon.pdf(x=rng, scale=lam), index=rng)
exp_cdf = pd.Series(data=expon.cdf(x=rng, scale=lam), index=rng)

geo_pmf = pd.Series(data=geom.pmf(k=rng, p=1/lam), index=rng)
geo_cdf = pd.Series(data=geom.cdf(k=rng, p=1/lam), index=rng)

pois_cdf = pd.Series(data=poisson.cdf(k=rng, mu=lam), index=rng)
```

```python
plt.figure(figsize=(20,7))
plt.plot(exp_pdf, label='exp_pdf', color='red', ls='--')
plt.plot(exp_cdf, label='exp_cdf', color='red')
plt.plot(geo_pmf, label='geo_pmf', color='blue', ls='--')
plt.plot(geo_cdf, label='geo_cdf', color='blue')
plt.plot(pois_cdf, label='pois_cdf', color='green')

plt.axhline(0.9, color='red', ls='--', lw=1)
plt.legend()
plt.show()
```
