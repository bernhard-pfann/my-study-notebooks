```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rand_var(n):
  """Return an array that is the outcome of a sequence of trials"""
  return np.random.choice(a=[0,n**2], size=n, p=[1-(1/n), 1/n])

def above_epsilon(arr, epsilon):
  """Returns probability that the outcome is above epsilon"""

  return sum(arr>epsilon)/arr.size
```

```python
lst_p = []
lst_mu = []

# iterate over different values of n
for n in range(1, 1000):

  arr = rand_var(n)
  p   = above_epsilon(arr, 0)
  mu  = np.mean(arr)

  lst_p.append(p)
  lst_mu.append(mu)
```

```python
plt.plot(lst_p)
plt.title('Probability > Epsilon')
plt.show()
```

```python
plt.plot(lst_mu)
plt.title('Expectation')
plt.show()
```
