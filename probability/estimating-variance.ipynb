## Convergence of Sample Mean
***

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p=0.3

# True variance when we know that X is Bernoulli
p*(1-p)
```

```python
# Variance from the sample data
np.var(np.random.binomial(n=1, p=p, size=100))
```

It can be seen that with increasing sample size, the derived $\hat \sigma$ converges to the true $\sigma$.

```python
var_X = []

# Calculate variance from sample data of different sizes
for size in range(100,100000,100):
  X = np.random.binomial(n=1, p=p, size=size)
  var_X.append(np.var(X))

plt.figure(figsize=(15,7))
plt.plot(var_X)
plt.title('Sample Variance for different n')
plt.show()
```
