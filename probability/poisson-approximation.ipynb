## Normal Approximation of Poisson
***

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def print_stats(n, p):
  print('Poisson Lambda:', n*p)
  print('Binom Mean:', np.mean(n*p))
  print('Binom Var:', n*p*(1-p))
```

Binomial $=$ Poisson $\not =$ Normal

```python
n=100
p=0.01
print_stats(n, p)

sns.histplot(x=np.random.binomial(n=n, p=p, size=10000), kde=True, color='b', alpha=0.2, lw=0, label='binom')
sns.histplot(x=np.random.poisson(lam=n*p, size=10000), kde=True, color='r', alpha=.2, lw=0, label='Poisson')
plt.legend()
plt.show()
```
