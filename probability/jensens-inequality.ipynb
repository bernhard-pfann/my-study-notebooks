```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
g = lambda x:x**2
X = np.arange(0, 11, 0.1)
g_X = pd.Series(data=g(X), index=X)
```

```python
a=3.0
b=8.0
p=0.5

c=a*p+(1-p)*b
c
```

```python
plt.figure(figsize=(15,7))
plt.plot(g_X,c='red')
plt.axvline(a, c='grey', lw=1, ls='--')
plt.axvline(b, c='grey', lw=1, ls='--')
plt.axvline(c, c='black', lw=1, ls='--')

plt.plot([a,b], [g(a), g(b)], c='blue')
```
