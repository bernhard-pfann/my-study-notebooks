## Central Limit Theorem
***

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

In this snippet we take $n$ random samples $X_i$ from a distributions, and sum up all $X_i$ to $S_n$. We repeat this process $10^6$ times to get an empirical density function.

```python
def uniform(n, iterate=10**6):
  return np.random.random(n*iterate).reshape(n, iterate).sum(axis=0)

def geometric(n, iterate=10**6):
  return np.random.geometric(p=0.2, size=n*iterate).reshape(n, iterate).sum(axis=0)
```

We can see that the larger $n$, the closer their sum gets to look like a normal. The QQ-plot is a measure of fit on a theoretical normal distribution.

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(25,10))

ax[0][0].hist(uniform(2), bins=20, edgecolor='w')
ax[0][1].hist(uniform(4), bins=20, edgecolor='w')
ax[0][2].hist(uniform(30), bins=20, edgecolor='w')

sm.qqplot(uniform(2), line='s', ax=ax[1][0])
sm.qqplot(uniform(4), line='s', ax=ax[1][1])
sm.qqplot(uniform(30), line='s', ax=ax[1][2])

ax[0][0].set_title('n=2')
ax[0][1].set_title('n=4')
ax[0][2].set_title('n=30')

plt.suptitle('Uniform Distribution', fontsize=16)
plt.show()
```

This is true no matter which distribution the individual $X_i$ have. We can see it in the case, where $X_i$ is geometric, but the sum of all $X_i$ still converges to a normal with $n \to \infty$. However, the convergence is slower the more asymmetric the underlying distribution is.

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(25,10))

ax[0][0].hist(geometric(2), bins=20, edgecolor='w')
ax[0][1].hist(geometric(4), bins=20, edgecolor='w')
ax[0][2].hist(geometric(30), bins=20, edgecolor='w')

sm.qqplot(geometric(2), line='s', ax=ax[1][0])
sm.qqplot(geometric(4), line='s', ax=ax[1][1])
sm.qqplot(geometric(30), line='s', ax=ax[1][2])

ax[0][0].set_title('n=2')
ax[0][1].set_title('n=4')
ax[0][2].set_title('n=30')

plt.suptitle('Geomtric Distribution', fontsize=16)
plt.show()
```
