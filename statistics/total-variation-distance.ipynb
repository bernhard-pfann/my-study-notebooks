## Total Variation Distance
***

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, uniform
```

```python
x = np.linspace(start=-0.1, stop=2, num=100000)

f_x = expon.pdf(x)
g_x = uniform.pdf(x)

plt.figure(figsize=(15,7))
plt.plot(x, f_x, c="red", label="exponential")
plt.plot(x, g_x, c="blue", label="uniform")

plt.legend()
plt.show()
```

```python
# computational solution
max(abs(g_x-f_x))
```

```python
# analytical solution
1-1/math.e
```
