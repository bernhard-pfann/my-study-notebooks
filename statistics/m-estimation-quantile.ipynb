## M-Estimation (Quantiles)
***

```python
import numpy as np
import pandas as pd
```

```python
# Quantile level
alpha = 0.95

# Observed data
X = np.random.randint(low=0, high=100, size=1000)
```

$$
C_\alpha =
\begin{cases}
    -(1-\alpha)x & \text{if } x <0\\
    \alpha x & \text{if }x \ge 0
\end{cases}
$$

```python
def loss_func(theta: int, X: np.array):
    """Loss function we want to minimize"""
    diff = X-theta
    return np.where(diff<0, -(1-alpha)*diff, alpha*diff).sum()
```

```python
# Try out different values for theta
theta_lst = pd.Series(np.arange(60,120,2))
loss = pd.Series({t: loss_func(theta=t, X=X) for t in theta_lst})
```

```python
import matplotlib.pyplot as plt

plt.plot(loss)
plt.show()
```
